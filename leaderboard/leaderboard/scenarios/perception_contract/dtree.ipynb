```python
import pandas as pd

dataset = pd.read_csv('out/dataset.data', header=None)
dataset1 = pd.read_csv('classified_dataset.csv', header=None)
```

```python
X = dataset.drop(columns=[9])
y = dataset[9]
X1 = dataset1.drop(columns=[2])
y1 = dataset1[2]
```

```python
# cat_idx = dataset['label'].unique().tolist()
cat_idx = []
```

```python
clf = Classifier(
    use_oblique=True,
    max_depth=-1,
    random_state=42,
    n_pair=X.shape[1],
    categories=cat_idx
)
```

```python
clf.fit(X1, y1)
```

```python
export_tree(clf, 'dtree123.json')
```

```python
for d in dataset.iterrows():
    # test decision tree
    dec = clf.predict(d[1].drop(9).values.reshape(1, -1))
    # print(f"Decision: {dec}, Expected: {d[1][9]}")
    if (dec[0] == 1) == False:
        print(d[1].drop(9).values.reshape(1, -1))
```
